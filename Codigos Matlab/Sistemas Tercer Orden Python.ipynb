```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
#Puede ser usado el "Python Control Systems Library" si se desea
```

```python
#Parámetros del sistema
Wn = 1;
ep = 0.7;
beta = [0.5, 1, 2, 4, 10];
```

## función de transferencia

$G=\dfrac{\omega_n^2p}{(s^2+2\zeta\omega_ns+\omega_n^2)(s+p)}$

```python
tm = np.linspace(0,20,100)
for i in beta:
    p = i*ep*Wn;
    num = [Wn**2*p];
    den = np.convolve([1, 2*ep*Wn, Wn**2], [1, p])
    G = signal.TransferFunction(num,den)
    t,y = signal.step(G,None,tm,20)
    plt.figure(1)
    plt.plot(t,y,label=r'$\beta=$'+str(i))
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.title('Step response')
plt.grid()
plt.legend()  
```
